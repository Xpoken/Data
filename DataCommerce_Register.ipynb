```python
# Aligner à droite ou a gauche les cellules d'une dataframe
def left_align(df):
    left_aligned_df = df.style.set_properties(**{'text-align': 'left'})
    left_aligned_df = left_aligned_df.set_table_styles(
        [dict(selector='th', props=[('text-align', 'left')])]
    )
    return left_aligned_df

def right_align(df):
    right_align_df = df.style.set_properties(**{'text-align': 'right'})
    right_align_df = right_align_df.set_table_styles(
        [dict(selector='th', props=[('text-align', 'right')])]
    )
    return right_align_df
```

```python
def CommerceSearchIceLand(SearchedNameString):
  # Company Registrar IceLand  Search Tool V0
  #https://www.skatturinn.is/fyrirtaekjaskra/leit

  # Webpage url     

  #Recherche Societe 
  # SearchedNameString = "eur"
  SearchedName = SearchedNameString
  url = f'https://www.skatturinn.is/fyrirtaekjaskra/leit?nafn={SearchedName}&heimili=&kt=&vsknr='
  print(url)

  # Extract tables
  dfs = pd.read_html(url)
  print(f"Nombre de tableau : {len(dfs)}")
  # Get first table                                                                                                           
  df = dfs[0]

  df.columns = ['RegistrationNumber', 'CompanyName', 'Address']

  # Note : (Félag afskráð)	 : deregistered    .str.replace('(Félag afskráð)', '(deregistered)', regex=False)
  df.CompanyName = df.CompanyName.str.replace('(Félag afskráð)', '(deregistered)', regex=False)
  df["UrlCompanyRegister"] = "https://www.skatturinn.is/fyrirtaekjaskra/leit/kennitala/"+df.RegistrationNumber.astype(str)	
  #df.style.format({'UrlCompanyRegister': make_clickable})

  #Rendre les liens clicable
  from IPython.core.display import display, HTML
  HTML(df.to_html(render_links=True, escape=False))

  #CurrentDate = pd.Timestamp.today()
  import datetime as dt    
  CurrentDate = dt.datetime.today().strftime("%Y-%m-%d")
  fileName=f"/content/Exports/{CurrentDate}_Company_register_Islande_{SearchedName}.xls"
  SheetName = f"{CurrentDate}_{SearchedName}"
  df.to_excel(fileName,
             sheet_name= SheetName) 
  #ExportDfColab(df,fileName)

  data = df.copy()
  data
  HTML(data.to_html(render_links=True, escape=False))
  return data

  #https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html
  #https://stackoverflow.com/questions/42263946/how-to-create-a-table-with-clickable-hyperlink-in-pandas-jupyter-notebook
  #https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_excel.html
```
